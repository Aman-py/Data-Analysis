```python
from matplotlib import pyplot as plt
x = [1,2,3,4,5,6]
y = [9,8,7,6,5,4]
x1 =[1,4,6,8,11]
y1 =[9,7,5,3,1]
plt.plot(x,y,'g',label='line1',linewidth=5)
plt.plot(x1,y1,'r',label='line2',linewidth=5,)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Test')
plt.grid(True,color='g')
plt.legend()
plt.show()
```

```python
#Bar Graph
from matplotlib import pyplot as plt
x = [1,2,3]
y = [9,8,7]
x1 =[4,6,5]
y1 =[10,11,12]
plt.bar(x,y,color='r')
plt.bar(x1,y1,color='g')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Test')
plt.show()
```

```python
#stack plot
from matplotlib import pyplot as plt
days=[1,2,3,4,5]
sleep=[1,2,1,2,1]
eating=[1,4,3,4,3]
working=[1,2,5,6,5]
playing=[7,8,7,8,7]
plt.stackplot(days,sleep,eating,working,playing,colors=['m','c','y','k'],labels=['sleep','eating','working','playing'],linewidth=5)
plt.legend()
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Day-Chart')
plt.show()
```

```python
#Pie Chart
from matplotlib import pyplot as plt
slices=[9,8,7,6]
acti=['sleep','eating','working','playing']
plt.pie(slices,labels=acti,startangle=90,colors=['r','m','y','b'],autopct='%1.1f%%',explode=[0,0,0,.1],shadow=True)
plt.title('Day-Chart')
plt.show()
```

```python
from matplotlib import pyplot as plt
from numpy import *
def f(t):
    return t**2+t%10
t1=array([1,2,3,4,5])
t2=array([6,7,8,9,10])
plt.subplot(411)
plt.plot(t1,f(t1),'bo',t2,f(t2))
plt.subplot(412)
plt.plot(t2,sin(t2*pi))
plt.subplot(413)
plt.plot(t1,log(t1))
plt.subplot(414)
plt.plot(t2,cos(t2))
plt.show()
```
